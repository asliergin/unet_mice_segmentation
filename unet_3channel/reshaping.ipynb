```python
import tifffile
import numpy as np
import re
import matplotlib.pyplot
import scipy.misc
import os
```

```python
#loading the tiff file into a numpy ndarray
image = tifffile.imread("4_G032_05-14-2019 17-02-26_4ms ICG clearance M30.tiff")
print(np.shape(image))

#reshaping the numpy ndarray to seperate 4 lasers
frames, height, width = image.shape
channels = 4
reshaped_image = image[:(frames//channels)*channels, ...].reshape((frames // channels, channels, height, width))
print(np.shape(reshaped_image))

# now we have (number of frames, number of lasers, height, width)
```

```python
#if we want to have seperate lasers:
laser1 = reshaped_image[:,0,:,:]
laser2 = reshaped_image[:,1,:,:]
laser3 = reshaped_image[:,2,:,:]
laser4 = reshaped_image[:,3,:,:]


print(np.shape(laser1))

```

```python
#create different directories to save png files that will be obtained from different lasers
path1 = "laser1"
path2 = "laser2"
path3 = "laser3"
path4 = "laser4"
path5 = "channels3" 
path6 = "channels4"

if not os.path.exists(path1):
    os.mkdir(path1)

if not os.path.exists(path2):
    os.mkdir(path2)

if not os.path.exists(path3):
    os.mkdir(path3)

if not os.path.exists(path4):
    os.mkdir(path4)
    
if not os.path.exists(path5):
    os.mkdir(path5)
    
if not os.path.exists(path6):
    os.mkdir(path6)
```

```python
current_path = os.getcwd()
```

```python
#if we want to combine 3 lasers in an image, and make it 3 channel
merge3lasers = reshaped_image[:,0:3,:,:] 
print(np.shape(merge3lasers))

```

```python
#if we want to combine 4 lasers in an image, and make it 4 channel
merge4lasers = reshaped_image[:,0:4,:,:] 
print(np.shape(merge4lasers))
```

```python
# Preprocessing: to be done, subtract mean
```

```python
#save images having 3 channels
for n,s in enumerate(merge3lasers):
    #add: if file doesn't exist
    transposed_s = np.transpose(s, (1, 2, 0))
    scipy.misc.imsave( current_path + "/channels3/" + str(n) + '_'+'channels3.png', transposed_s)
```

```python
#save images having 4 channels
for n,s in enumerate(merge4lasers):
    #add: if file doesn't exist
    transposed_s = np.transpose(s, (1, 2, 0))
    scipy.misc.imsave( current_path + "/channels4/" + str(n) + '_'+'channels4.png', transposed_s)
```

```python
#save images from laser1
for n,s in enumerate(laser1):
    #add: if file doesn't exist
    scipy.misc.imsave( current_path + "/laser1/" + str(n) + '_'+'laser1.png', s)
```
